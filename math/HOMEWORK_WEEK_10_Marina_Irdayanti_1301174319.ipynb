# Exercise

 Diberikan PDP dengan nilai awal dan batas seperti berikut:
\begin{align*}
\frac{\partial u}{\partial t} &= \mu \frac{\partial^2 u}{\partial x^2}, \quad x \in (0,1), \quad t>0\\
 u(x,0)&=f(x), \quad x \in [0,1]\\
 u(0,t)&=a(t), \quad u(1,t)=b(t), \quad t\geq 0
\end{align*}
dengan $f(0)=a(0)$, dan $f(1)=b(0)$.


*   Tentukan nilai $v(k\Delta x, n\Delta t)$ untuk $n=0,1,2,\cdots 5$, dengan $f(x)=\sin \frac{\pi x}{2}$, $a=b=0$, $M=20$, $\mu=1$, dan $\Delta t=1.2 \times 10^{-3}$ menggunakan program komputer!
*   Sama dengan pertanyaan (a), akan tetapi gunakan $\Delta t=1.3 \times 10^{-3}$!



# **Bagian: Δt=1.2×10−3**

```
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return np.sin(np.pi*x/2)

L = 1
M = 20
mu = 1 
dt = 1.2*10**(-3)
T = 5*dt
a = 0 
b = 0

dx = L/M 

r = mu*dt/(dx*dx)


x=[]
vold = []
for k in range(0,M+1):
  x.append(k*dx)
  vold.append(f(x[k]))

plt.plot(x,vold)
plt.draw()
```

```
time =0
while (time < T):
  time = time + dt
  v = []
  v.append(a)
  for k in range(1,M):
    v.append(vold[k]+r*(vold[k+1]- 2*vold[k] +vold[k-1]))
  
  v.append(b)
  
  vold = v

  plt.plot(x,vold)
  plt.draw()
```
