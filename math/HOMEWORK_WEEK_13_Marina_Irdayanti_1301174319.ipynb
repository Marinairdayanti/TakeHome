# **Exercise (4.2.10), t = 0.06 and dt = 0.0005**

---



```
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt


def nu(x, y):
    return 1*((x**2) + (y**2)) 

Lx =  1
Ly =  1

Nx =  20
Ny =  20

D = 1

x = np.linspace(0,Lx,Nx) #for plot purpose
y = np.linspace(0,Ly,Ny) #for plot purpose
X,Y = np.meshgrid(x,y)
dx =  x[1] - x[0]
dy =  y[1] - y[0] 
print("dx = ",dx)
print("dy = ",dy)

T =  0.06
Tfin = 0.25  
dt = 0.0005
print("dt = ",dt)



#Initial condition
u = np.zeros((Nx,Ny))
u_n = np.zeros((Nx,Ny))

#setting initial condition u
for i in range(0,Nx):
  for j in range(0,Ny):
    u[i][j] = nu(x[i],y[j]) 

#print(rho)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, u, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()

cp = plt.contourf(X, Y, u)
plt.colorbar(cp)
plt.show()
```

```
time =0 
while(time <=Tfin):
  time = time + dt 

  for i in range(1,Nx-1):
   for j in range(1,Ny-1):
     u_n[i][j] = u[i][j] + (D*dt/dx**2)*(u[i+1][j] - 2*u[i][j] + u[i-1][j]) + (D*dt/dy**2)*(u[i][j+1] - 2*u[i][j] + u[i][j-1])
  
  #boundary
  for i in range(0,Nx):
     u_n[i][0] = 0
     u_n[i][Ny-1] = 0
  for j in range(0,Ny):
     u_n[0][j] = 0
     u_n[Nx-1][i] = 0
  u = u_n

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, u, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set(xlim=(0, Lx), ylim=(0, Ly), zlim=(0,50))
plt.show()

cp = plt.contourf(X, Y, u)
plt.colorbar(cp)
plt.show()
```
