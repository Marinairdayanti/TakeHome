```
import numpy as np

def F(X):
  f1 = 3*X[0]**2 - X[1]**2 
  f2 = 3*X[0]*X[1]**2 - (X[0]**3) - 1
  
  return np.array([f1, f2])

def g(X):
  fx = F(X)
  return  fx[0]**2 + fx[1]**2 

def Grad_g(X):
  fx = F(X)

  f1x = 2*fx[0]*(6*X[0]) + 2*fx[1]*((3*X[1]**2)-3*X[0]) 
  f2x = 2*fx[0]*(2*X[1]) + 2*fx[1]*(6*X[0]) 
  
  return np.array([f1x, f2x])

def errorL2(X):
  return  np.sqrt(sum(X**2))

x=[1,1]
TOL = 0.01
N = 100
k = 1

#step 1
k= 1
while (k<=N):   #step 2
  #step 3
  g1 =  g(x)
  Z =  Grad_g(x)
  z0 =  errorL2(Z)
  
  #step 4
  if (z0 ==0):
    print("Zero gradient")
    print(x)
    break;
  #Step 5
  Z  =  Z/z0
  a1 =  0
  a3 =  1
  g3 = g(x-a3*Z) 
  #print(g3)

  #step 6
  while(g3 >= g1):
    #step 7
    a3 =  a3/2
    g3 =  g(x-a3*Z) 
    #Step 8
    if a3 < TOL/2:
      print("No likely improvement")
      print(x)
      break
  #step 9
  a2 =  a3/2
  g2 =  g(x-a2*Z) 

  #step 10
  h1 =  (g2-g1)/a2
  h2 =  (g3-g2)/(a3-a2)
  h3 =  (h2-h1)/a3

  #step 11
  a0 =  0.5*(a2-h1/h3)
  g0 =  g(x-a0*Z) 

  #step 12
  if (g0<g3):
    a = a0
  else:
    a = a3

  gL= g(x-a*Z)
  #Step 13
  x = x - a*Z
  print("{}, {}, {}".format(k,x,g(x)))
  #Step 14
  if abs(gL-g1) < TOL:
    print("done")
    print(x)
    break
  
  k=k+1
```
